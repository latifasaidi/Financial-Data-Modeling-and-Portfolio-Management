```python
import numpy as np
overbuy=70
neutral_buy=55
#put nan value for signal long column
f["signal_long"]=np.nan
f["yesterday"]=f['rsi'].shift(1)
#define the open long signal
f.loc[(f['rsi']> neutral_buy) & (f['yesterday']< neutral_buy), "signal_long"]=1
#define the close long signal: signal fausse
f.loc[(f['rsi']< neutral_buy) & (f['yesterday']> neutral_buy), "signal_long"]=0
#define the close long signal: sur achat 
f.loc[(f['rsi']< overbuy) & (f['yesterday']> overbuy ), "signal_long"]=0
```

```python
#determiner le pourcentage de la variation d' asset
f['pct']= f['Adj Close'].pct_change(1)

#determiner le return de la stratégie 
f["return"]= f["pct"] * f["position"].shift(1)
f["return"].loc["2010"].cumsum().plot()
```

```python
def RSI( val, neutral,windo):

  if "Adj Close" not in val.columns:
   ValueError("we need a column named'adj close' for continuing the computing process")
  val['rsi']=ta.momentum.RSIIndicator(f["Adj Close"], window=windo).rsi()

  overbuy=70
  neutral_b=50+ neutral

  val["signal_long"]=np.nan
  val["yesterday_rsi"]= f["rsi"].shift(1)

  val.loc[(val["rsi"]> neutral_b) & (val["yesterday_rsi"]< neutral_b), "signal_long"]=1
  val.loc[(val["rsi"]< neutral_b) & (val["yesterday_rsi"]> neutral_b), "signal_long"]=0
  val.loc[(val["rsi"]< overbuy) & (val["yesterday_rsi"]> overbuy), "signal_long"]=0

  oversell=30
  neutral_s=50- neutral
  val["signal_short"]=np.nan
  val["yesterday_rsi"]= f["rsi"].shift(1)

  val.loc[(val["rsi"]< neutral_b) & (val["yesterday_rsi"]>neutral_b), "signal_short"]=-1
  val.loc[(val["rsi"]> neutral_b) & (val["yesterday_rsi"]< neutral_b), "signal_short"]=0
  val.loc[(val["rsi"]>overbuy) & (val["yesterday_rsi"]< overbuy), "signal_short"]=0

  val["pct"]=val["Adj Close"].pct_change(1)
  val["position"]= (val["signal_short"].fillna(method="ffill")) + (val["signal_long"].fillna(method="ffill"))
  val["return"]= val["pct"]*val["position"].shift(1)
  return val["return"]
```

```python
#sortino<0: investissement non rentable (gain negatif)
#sortino<1: "investissement rentable mais risques superieurs aux gains"
#sortin >1: tres bon investissement (  risques inferieurs aux gains)
ft=yf.download("GOOG", end= "2021-01-01")
return_series=ft["Adj Close"].pct_change(1).dropna()
return_series.name ="return"
return_series
moy=np.mean(return_series)
vol= np.std(return_series[return_series<0])
sortino= np.sqrt(252) * moy/ vol
print(f"sortin= {'%.3f' %sortino}")
```

```python
#beta<1: Bon: portfeuille a une variation moindre que l'indice (si l'indice varie de 1 le portfeuille varie de valeur=beta)
#beta>1: n'est pas bon: la variation de portfeuille depasse l'indice


sp500=yf.download("^GSPC")["Adj Close"].pct_change(1)
sp500.name="SP500"
val= pandas.concat((return_series, sp500), axis=1).dropna()
M_var_cov= np.cov(val.values, rowvar=False)
M_var_cov
cov=M_var_cov [0] [1] #ou bien 1 0 car cov_var est symétrique 
var= M_var_cov[1][1]
beta= cov/ var
print(f"beta est egal:{'%.5f' %beta} ")
```

```python
#alpha>0: le portfeuille surperforme le marché en terme de rendement risque
#alpha<0: le portfeuille sous performe le marché en terme de rendement risque
 #calcul of alpha
alpha= (252*moy*(1-beta))*100
print(f"alpha est egal= {'%.3f' %alpha} %")
```

```python
def drawdown_function (serie):
  #on calcul le cumulé
  cum=serie.dropna().cumsum()+1
  #calculer le max pendant une periode: methode:(1,2,1,5,8,3,4,7)--->(1,2,1,5,8,8,8,8)
  running_max=np.maximum.accumulate(cum)
  #calculer le drawdown
  return(cum/running_max -1)
```

```python
def backTest(serie):
  #import the benchmark 
  sp500=yf.download("^GSPC")["Adj Close"].pct_change(1).loc["2019-01-01":"2021-01-01"]
   #change the name
  sp500.name="SP500"

   #Concatenate Tthe RETURNS and the SP500
  val=pandas.concat((return_series, sp500), axis=1).dropna()

   #compute the drawdown
  drawdown=drawdown_function(serie)

   #the max drawdown
  max_drawdown= - np.min(drawdown)

   #put a subplot
  fig, (cum, dra)= plt.subplots(1, 2, figsize=(20, 6))

   #put the suptitle
  plt.suptitle("BackTesting", size=20)

   #return cusm chart
  cum.plot(serie.cumsum()*100, color="#39B3C7")
  cum.plot(val['SP500'].cumsum()*100, color="#BB5A0F")

   #put a legend
  cum.legend(["portfolio", "SP500"] )

   #set individual title
  cum.set_title("cumulative return", size=13)
  cum.set_ylabel("cumulative return %", size=11)

   #put the rawdown
  dra.fill_between(drawdown.index, 0, drawdown, color="#C73954",alpha=0.65 )

   #set individual title
  dra.set_title("drawdown", size=13)
  dra.set_ylabel("drawown en  %", size=11)
   #plot the graph
  plt.show()

   #compute the sortino
  sortino= np.sqrt(252)* serie.mean() / serie.loc[serie<0].std()

   #compute the beta///////
  beta= np.cov( val[["return", "SP500"]].values, rowvar=False) [0] [1] / np.var (val["SP500"].values)
  #compute the alpha
  alpha= 252*serie.mean() - beta*serie.mean()* 252

  #print the statistics
  print("sortino: ", np.round(sortino, 3))
  print(f"beta : { np.round(beta, 3)}")
  print(f"alpha : { np.round(alpha*100, 3)}%")
  print(f"drawdown maximum: { np.round(max_drawdown * 100, 3)} %")
backTest(return_series)
```

```python
def grid_parameters(f):
  neutral_values=[i*2 for i in range(10)]
  window_values= [i*2 for i in range(1, 10)]

  grid=np.zeros((len(neutral_values), len(window_values)))
  for i in range(len(neutral_values)):
    for j in range(len(window_values)):
      return_rsi=RSI(f, neutral_values[i], window_values[j])
      grid[i][j]= np.sqrt(252)*return_rsi.mean() / (return_rsi[return_rsi<0]).std()
  return(grid)
```
